```python
import pandas as pd

# Assuming the data is already loaded into a DataFrame called 'df'

# Define a list of common two-wheeler brands
motorcycle_brands = ['Royal Enfield', 'UM', 'KTM', 'Bajaj', 'Hyosung', 'Mahindra', 'Honda', 'Yamaha', 'TVS', 'Hero', 'Suzuki', 'Activa']

# Filter the DataFrame to include only motorcycles
motorcycles = df[df['Car_Name'].str.contains('|'.join(motorcycle_brands), case=False, na=False)]

# Find the oldest motorcycle
oldest_motorcycle = motorcycles.loc[motorcycles['Year'].idxmin()]

print("Number of motorcycles:", len(motorcycles))
print("\nOldest motorcycle:")
print(oldest_motorcycle)
```

```python
# Find the newest motorcycle
newest_motorcycle = motorcycles.loc[motorcycles['Year'].idxmax()]

print("Number of motorcycles:", len(motorcycles))
print("\nNewest motorcycle:")
print(newest_motorcycle)
```

```python
most_sold_bike = motorcycles['Car_Name'].mode().iloc[0]
print("Most sold bike:", most_sold_bike)
```

```python
# Assuming your DataFrame is named 'df'

# Convert the 'Year' column to datetime format (assuming it contains years)
df['Year'] = pd.to_datetime(df['Year'], format='%Y')

# Sort the DataFrame by the 'Year' column in ascending order (oldest to newest)
df_sorted = df.sort_values('Year')

# Find the oldest car (first row)
oldest_car = df_sorted.iloc[0]

# Find the newest car (last row)
newest_car = df_sorted.iloc[-1]

print("Oldest car:")
print(oldest_car)

print("\nNewest car:")
print(newest_car)


```

```python
# Assuming you've already loaded the data into a DataFrame called 'df'
df['Price_Ratio'] = df['Selling_Price'] / df['Present_Price']
top_deals = df.nlargest(10, 'Price_Ratio')
print(top_deals)
```

```python
# Calculate the ratio of Selling_Price to Present_Price
df['Price_Ratio'] = df['Selling_Price'] / df['Present_Price']

# Sort the DataFrame by Price_Ratio in descending order to find the top deals
top_deals = df.sort_values(by='Price_Ratio', ascending=False).head(10)

# Display the top 10 deals
print("Top 10 Deals:")
print(top_deals[['Car_Name', 'Price_Ratio']])

# Theoretical reasons for high resale value of these cars
print("\nReasons for high resale value of these cars:")
print("1. Low Mileage: Cars with low kilometers driven indicate less wear and tear.")
print("2. Recent Model Years: Newer cars often retain more of their original value.")
print("3. Dealer Sales: Dealers usually offer well-maintained and certified vehicles.")
print("4. Popular Models: Cars known for reliability and high demand maintain strong resale value.")
print("5. Fuel Type: Popular fuel types like petrol and diesel are generally more in demand.")
```
